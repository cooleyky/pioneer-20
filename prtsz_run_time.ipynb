```python
# Import libraries used in this note
import os

import pandas as pd
```

```python
# Load data from CTD survey PRTSZ experiment
exp_file = os.path.join(os.path.expanduser("~/Desktop"),
                        "pioneer-20_data/PRTSZ_on_rosette",
                        "AR82b_CTDCrossShelfSurvey_LISSTBatteryExperiment.xlsx")
prtsz_exp = pd.read_excel(exp_file)
prtsz_exp
```

```python
time_start = pd.to_datetime(prtsz_exp["Sampling start time"].drop([0, 1, 3, 4, 6, 8, 10, 12, 14, 16, 17, 19, 21, 23, 25, 27]))
time_start
```

```python
time_stop = pd.to_datetime(prtsz_exp["Sampling stop time"].drop([0, 1, 3, 4, 6, 8, 10, 12, 14, 16, 17, 19, 21, 23, 25, 27]))
time_stop
```

```python
operational_time = time_stop-time_start + pd.Timedelta("00:00:01")
operational_time
```

```python
# Find cummulative run time from one battery
cummulative_run_time = operational_time.iloc[0:6].sum()
```

```python
cummulative_run_time.seconds
```
